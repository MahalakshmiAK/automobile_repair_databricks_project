```python
%sql
--Average days a vehicle spends in the shop, by store and service type.
SELECT 
    fi.store_id,  service_type,vehicle_in_date, vehicle_out_date , 
    ROUND(AVG(DATEDIFF(vehicle_out_date, vehicle_in_date)) 
        OVER(PARTITION BY fi.store_id, service_type ORDER BY fi.store_id,service_type),2) AS Avg_Days_In_Shop
FROM workspace.gold.fact_invoice_details fi JOIN workspace.gold.dim_order o 
ON fi.order_id = o.order_id WHERE order_status="COMPLETED" ;
```
