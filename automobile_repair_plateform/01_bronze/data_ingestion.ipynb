```python
import pyspark.sql.functions 

tables=["customer_survey","invoice","order","estimate","ns_budget","store"]
for table in tables:
    source_table_name="workspace.azure_blob_storage."+table
    df = spark.read.table(source_table_name)
    destination_table_name="workspace.bronze."+table
    df.write.mode("overwrite").saveAsTable(destination_table_name)
```
